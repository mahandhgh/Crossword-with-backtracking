```python
lines_of_puzzle, length_of_lines = map(int, input().split()) #The number of pattern lines and length of this lines
pattern_structure = [list(''.join(input())) for i in range(lines_of_puzzle)] # Make a pattern structure | Note : please enter characters without white space!
words_list = input().split() # Make a list of words

# An example:

# pattern_structure = [
#     ['#', '_', '#', '#', '#', '#', '_', '_', '_', '_', '_', '#'],
#     ['#', '_', '#', '#', '#', '#', '_', '#', '#', '#', '#', '#'],
#     ['#', '_', '#', '#', '_', '_', '_', '_', '_', '#', '#', '#'],
#     ['#', '_', '#', '#', '#', '#', '_', '#', '#', '#', '#', '#'],
#     ['_', '_', '_', '_', '_', '_', '_', '_', '_', '#', '#', '#'],
#     ['#', '#', '_', '#', '#', '#', '_', '#', '#', '_', '#', '#'],
#     ['#', '#', '_', '#', '_', '#', '_', '_', '_', '_', '_', '_'],
#     ['#', '_', '_', '_', '_', '_', '#', '#', '#', '_', '#', '#'],
#     ['#', '#', '#', '#', '_', '#', '#', '#', '#', '_', '#', '#'],
#     ['#', '#', '#', '#', '_', '#', '#', '#', '#', '#', '#', '#'],
#     ['#', '#', '#', '#', '_', '#', '#', '#', '#', '#', '#', '#']
# ]

# words_list = ["queue", "stack", "array", "sorting", "recursion", 
#               "char", "greedy", "sets", "trees", "deque"]

```

```python
def can_place_word_horizontally(puzzle, word, row, col):
    if col + len(word) > len(puzzle[0]):
        return False
    for j in range(len(word)):
        if puzzle[row][col+j] not in ['_', word[j]]:
            return False
    return True

def place_word_horizontally(puzzle, word, row, col):
    original = []
    for j in range(len(word)):
        original.append(puzzle[row][col+j])
        puzzle[row][col + j] = word[j]
    return original

def can_place_word_vertically(puzzle, word, row, col):
    if row + len(word) > len(puzzle):
        return False
    for i in range(len(word)):
        if puzzle[row+i][col] not in ['_', word[i]]:
            return False
    return True

def place_word_vertically(puzzle, word, row, col):
    original = []
    for i in range(len(word)):
        original.append(puzzle[row+i][col])
        puzzle[row+i][col] = word[i]
    return original

def remove_word_horizontally(puzzle, original, row, col):
    for j in range(len(original)):
        puzzle[row][col+j] = original[j]

def remove_word_vertically(puzzle, original, row, col):
    for i in range(len(original)):
        puzzle[row+i][col] = original[i]

def solve_the_crossword(puzzle, words):
    if not words:
        return True
    
    word = words[0]
    for i in range(len(puzzle)):
        for j in range(len(puzzle[0])):
            if can_place_word_horizontally(puzzle, word, i, j):
                original = place_word_horizontally(puzzle, word, i, j)
                if solve_the_crossword(puzzle, words[1:]):
                    return True
                remove_word_horizontally(puzzle, original, i, j)
            
            if can_place_word_vertically(puzzle, word, i, j):
                original = place_word_vertically(puzzle, word, i, j)
                if solve_the_crossword(puzzle, words[1:]):
                    return True
                remove_word_vertically(puzzle, original, i, j)
    
    return False

def print_puzzle(puzzle):
    for row in puzzle:
        print(''.join(row))

def start(pattern, lst):
    if solve_the_crossword(pattern, lst):
        print("Crossword solved!")
        print_puzzle(pattern)
    else:
        print("The crossword cannot be solved!")

start(pattern_structure, words_list)
```
